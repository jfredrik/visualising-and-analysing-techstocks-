```python
import pandas as pd
import numpy as np
```

```python
import pandas_datareader as web
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
symbols = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'TSLA']

start_date = '2020-01-01'
end_date = '2022-10-01'

stock_data = web.get_data_yahoo(symbols, start_date, end_date)

stock_data
```

```python
stock_data['Adj Close']
```

```python
adj_closes = stock_data['Adj Close']
adj_closes.plot()
plt.title('Adjusted Closes Prices of Top 5 Tech Stocks by Market Cap from 2020 to 1st oct 2022, as of date 5th Oct')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Prices')
plt.show()
```

```python
daily_return = adj_closes.pct_change()
daily_return.plot()
plt.title('Daily Rate of Return')
plt.xlabel('Date')
plt.ylabel('Rate of Return')
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))

#AAPL
ax1 = plt.subplot(2, 3, 1)
plt.plot(daily_return['AAPL'], color='blue')
plt.title('APPL - Apple')
plt.xlabel('Date')
plt.ylabel('Daily Rate of Return')

#MSFT
ax2 = plt.subplot(2, 3, 2)
plt.plot(daily_return['MSFT'], color='orange')
plt.title('MSFT - Microsoft')
plt.xlabel('Date')
plt.ylabel('Daily Rate of Return')

#GOOG
ax3 = plt.subplot(2, 3, 3)
plt.plot(daily_return['GOOG'], color='green')
plt.title('GOOG - Google')
plt.xlabel('Date')
plt.ylabel('Daily Rate of Return')

#AMZN
ax4 = plt.subplot(2, 3, 4)
plt.plot(daily_return['AMZN'], color='red')
plt.title('AMZN - Amazon')
plt.xlabel('Date')
plt.ylabel('Daily Rate of Return')

#TSLA
ax5 = plt.subplot(2, 3, 5)
plt.plot(daily_return['AAPL'], color='purple')
plt.title('TSLA - Tesla')
plt.xlabel('Date')
plt.ylabel('Daily Rate of Return')

plt.subplots_adjust(wspace=0.3, bottom=0.3)

plt.show()
```

```python
mean_daily_return = daily_return.mean()

ax6 = plt.subplot()
ax6.set_xticks(range(len(symbols)))
ax6.set_xticklabels(symbols)

plt.bar(range(len(symbols)), mean_daily_return, color = 'black')

plt.xlabel('Stocks')
plt.ylabel('Mean Daily Rate of Return')
plt.title('Mean Rate of Return for APPLE, MSFT, GOOG, AMZN, TLSA')

plt.show()
```

```python
variance_daily_return = daily_return.var()

ax7 = plt.subplot()
ax7.set_xticks(range(len(symbols)))
ax7.set_xticklabels(symbols)

plt.bar(range(len(symbols)), variance_daily_return, color = 'black')

plt.xlabel('Stocks')
plt.ylabel('Variance')
plt.title('Variance of Stocks')

plt.show()
```

```python
std_daily_return = daily_return.std()

ax8 = plt.subplot()
ax8.set_xticks(range(len(symbols)))
ax8.set_xticklabels(symbols)

plt.bar(range(len(symbols)), std_daily_return, color = 'black')

plt.xlabel('Stocks')
plt.ylabel('Standard Deviation')
plt.title('Standard Deviation x Stocks')

plt.show()
```

```python
correlation_daily_return = daily_return.corr()
print(correlation_daily_return)
```
